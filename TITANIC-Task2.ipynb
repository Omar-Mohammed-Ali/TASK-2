```python
# importing libraries and loading the dataset
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
# Exploring the dataset
df.head()
df.info()
df.isnull().sum()
df.describe(include='all')
```

```python
# Handling missing values using imputations
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns = ['Cabin'])
df.isnull().sum()
```

```python
# Encode Categorical Features
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
df.drop(['Name', 'Ticket'], axis=1, inplace=True)
```

```python
# Normalizing numerical features
Scaler = StandardScaler()
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']
df[numerical_cols] = Scaler.fit_transform(df[numerical_cols])
```

```python
# Visualization and removing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numerical_cols])
plt.title("Boxplot Before Removing Outliers")
plt.show()
```

```python
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df
df_cleaned = remove_outliers(df, numerical_cols)
```

```python
# after removing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned[numerical_cols])
plt.title("Boxplot After Removing Outliers")
plt.show()
```

```python
# Cleaned Dataset
print("Final dataset shape after cleaning:", df_cleaned.shape)
```
